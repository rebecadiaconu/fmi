# Reprezentarea grafica a programelor liniare

## Avem urmatoarele cazuri:
1. incompatibil
2. compatibil cu o singura solutie optima
3. compatibil cu optim infinit
4. compatibil cu o infinitate de solutii optime

```python
import numpy as np
import matplotlib.pyplot as plt

def function(x, y):
    return x + y

# Reprezentam grafic
x = np.linspace(-5, 8, 200)
```

## I. Program liniar incompatibil

Dorim max(x + y), tinand cont de urmatoarele conditii:
- -2 * x + 2 * y >= 4 <==> -x + y >= 2
- x - y >= 0
- x > =0, y >= 0

```python
plt.figure(dpi=100)
plt.suptitle("Program liniar incompatibil")

# -x + y = 2
y = 2 + x
plt.plot(x, y)
plt.fill_between(x, y, y + 10, alpha=0.2)

# x - y = 0
y = x
plt.plot(x, y)
plt.fill_between(x, y, y - 7, alpha=0.2)

# adaugam axele OX, OY
plt.axvline(0, c='black')  # Adauga axa OY
plt.axhline(0, c='black')  # Adauga axa OX
plt.xlabel('x')
plt.ylabel('y')

plt.ylim(-1, 5)
plt.xlim(-5, 5)

plt.show()
```

### Explicatie grafic
- x - y >= 0 ==> x >= y, deci se vor lua in considerare doar valorile de sub graficul functiei.
-  -2x + 2y >= 4 <==> y >= 2 + x
<br>

Pentru a evita cazurile precum y = 0, x = -1: 0 >= 1 FALS, vom lua
in considerare doar valorile din partea de sus a graficului.
<br>

### Concluzie
Putem observa ca cele 2 semiplane nu se intersecteaza, deci nu avem unde cauta valoarea maxima
 a lui x + y.

<br>


## II. Program liniar cu optim unic

Dorim max(x + y), tinand cont de urmatoarele conditii
- x + 2 * y <= 4
- x - y >= 1
- x >= 0, y >= 0
<br>


```python
plt.figure(dpi=100)
plt.suptitle("Program compatibil cu optim unic")

# x + 2y = 4
y = (4 - x) / 2
plt.plot(x, y)
plt.fill_between(x, y, where=(x + 2 * y <= 4), alpha=0.3)

# x - y = 1
y = x - 1
plt.plot(x, y)
plt.fill_between(x, y, where=(x - y >= 1), alpha=0.2)

# adaugam axele OX, OY
plt.axvline(0, c='black')  # Adauga axa OY
plt.axhline(0, c='black')  # Adauga axa OX
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(1, 0, marker='x')
plt.scatter(2, 1, marker='x')
plt.scatter(4, 0, marker='x')
plt.show()
```

### Explicatie grafic
Observam ca cele 2 semiplane se intersecteaza in cele 3 puncte: A(1, 0), B(2, 1), C(4, 0) =>
 solutia optima unica a ecuatiei max(x + y) este punctul C(4, 0): 4 + 0 = 4 >= 2 + 1 = 3 >= 1 + 0 = 1
<br>
<br>

## III. Program liniar cu optim unic, infinit
Dorim max(x + y), tinand cont de urmatoarele conditii:
- x + y <= 4
- -4 * x + 2 * y <= 8
- x >= 0, y >= 0

```python
plt.figure(dpi=100)
plt.suptitle("Program compatibil cu optim unic")

# x + y = 4
y = 4 - x
plt.plot(x, y)
plt.fill_between(x, y, where=(x + y <= 4), alpha=0.3)

# -4x + 2y <= 8
y = (8 + 4 * x) / 2.0
plt.plot(x, y)
plt.fill_between(x, y, where=( -4 * x + 2 * y <= 8), alpha=0.2)

# adaugam axele OX, OY
plt.axvline(0, c='black')  # Adauga axa OY
plt.axhline(0, c='black')  # Adauga axa OX
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(0, 4, marker='x', color='red')
plt.scatter(4, 0, marker='x', color='blue')
plt.scatter(-2, 0, marker='x')
plt.show()
```

### Explicatie grafic
Observam ca cele 2 semiplane se intersecteaza in cele 3 puncte: A(4, 0), B(0, 4), C(-2, 0),
dar cum noi vom cauta solutii doar in primul cadran (x, y >= 0) =>
solutia optima unica a ecuatiei max(x + y) este intr-unul din punctele: A(4, 0), B(0, 4)

A(4, 0): 4 + 0 = 4
B(0, 4): 0 + 4 = 4, deci solutia ecuatiei se afla oriunde pe [A, B] si aplicand formula obtinem ca este de forma:
<br>
S = {4*lambda, 4 - 4*lambda | lambda din [0, 1]}
<br>

## IV. Program liniar cu optim infinit
Dorim max(x + y), tinand cont de urmatoarele conditii:
- x + 2 * y >= 2
- x >= 0, y >= 0

```python
plt.figure(dpi=100)
plt.suptitle("Program compatibil cu optim unic")

# x + 2y = 2
y = (2 - x) / 2
plt.plot(x, y)
plt.fill_between(x, y, y + 10, alpha=0.3)

# adaugam axele OX, OY
plt.axvline(0, c='black')  # Adauga axa OY
plt.axhline(0, c='black')  # Adauga axa OX
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(0, 1, marker='x', color='red')
plt.scatter(2, 0, marker='x', color='blue')
plt.show()
```

### Explicatie grafic
Observam ca cele 2 semiplane se intersecteaza in cele 2 puncte: A(2, 0), B(0, 1),
dar cum noi vom cauta solutii doar in primul cadran (x, y >= 0) =>
solutia optima unica a ecuatiei max(x + y) reprezinta intreaga suprafata hasurata a cadranului I.
